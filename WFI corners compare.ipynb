```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16, 6)  # make the default figure size larger
matplotlib.rcParams['image.interpolation'] = 'nearest'  # don't blur/smooth image plots
from matplotlib import pyplot as plt
import webbpsf
from webbpsf import wfirst
```

```python
webbpsf.setup_logging()
```

```python
wfi = wfirst.WFI()
wfi.filter = 'J129'
wfi.detector = 'SCA09'
wfi.detector_position = (4, 4)
psf_sca09 = wfi.calcPSF()
wfi.detector = 'SCA17'
wfi.detector_position = (4092, 4092)
psf_sca17 = wfi.calcPSF()
webbpsf.display_PSF_difference(psf_sca09, psf_sca17)
```

```python
webbpsf.display_PSF_difference(psf_sca09, psf_sca17, vmax=5e-3, title='', imagecrop=2.0)
```

```python
webbpsf.display_PSF(psf_sca09)
```

```python
webbpsf.display_PSF(psf_sca17)
```

```python
fig, (ax_sca09, ax_sca17, ax_diff) = plt.subplots(1, 3, figsize=(16, 4))
webbpsf.display_PSF(psf_sca09, ax=ax_sca09, imagecrop=2.0, title='WFI SCA09, bottom left - J129')
webbpsf.display_PSF(psf_sca17, ax=ax_sca17, imagecrop=2.0, title='WFI SCA17, top right - J129')
webbpsf.display_PSF_difference(psf_sca09, psf_sca17, vmax=5e-3, title='(SCA09) - (SCA17)', imagecrop=2.0, ax=ax_diff)
plt.savefig('./compare_wfi_sca09_sca17.png', bbox_inches='tight')
```

```python
fig, (ax_sca09, ax_sca17, ax_diff) = plt.subplots(1, 3, figsize=(16, 4))
webbpsf.display_PSF(psf_sca09, ext='DET_SAMP', ax=ax_sca09, imagecrop=2.0, title='WFI SCA09, bottom left - J129')
webbpsf.display_PSF(psf_sca17, ext='DET_SAMP', ax=ax_sca17, imagecrop=2.0, title='WFI SCA17, top right - J129')
webbpsf.display_PSF_difference(psf_sca09, psf_sca17, ext1='DET_SAMP', ext2='DET_SAMP', vmax=5e-2, title='(SCA09) - (SCA17)', imagecrop=2.0, ax=ax_diff, cmap='RdBu_r')
plt.savefig('./compare_wfi_sca09_sca17.png', bbox_inches='tight')
```
